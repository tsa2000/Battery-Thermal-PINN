```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from src.model import ThermalPINN

# -------------------------------
# 1️⃣ Load trained model
# -------------------------------
model = ThermalPINN(layers=[3, 64, 64, 64, 64, 5], activation='tanh')
model.load_state_dict(torch.load("results/battery_thermal_final.pth"))
model.eval()

# -------------------------------
# 2️⃣ Generate a grid of points inside the battery
# -------------------------------
x_min, x_max = 0.0, 0.1
y_min, y_max = 0.0, 0.05
z_min, z_max = 0.0, 0.01

nx, ny, nz = 20, 10, 5  # grid resolution
x = np.linspace(x_min, x_max, nx)
y = np.linspace(y_min, y_max, ny)
z = np.linspace(z_min, z_max, nz)

X, Y, Z = np.meshgrid(x, y, z)
points = np.vstack([X.ravel(), Y.ravel(), Z.ravel()]).T
points_tensor = torch.tensor(points, dtype=torch.float32)

# -------------------------------
# 3️⃣ Predict with the model
# -------------------------------
with torch.no_grad():
    predictions = model(points_tensor)

u = predictions[:, 0].numpy().reshape(X.shape)
v = predictions[:, 1].numpy().reshape(X.shape)
w = predictions[:, 2].numpy().reshape(X.shape)
T = predictions[:, 4].numpy().reshape(X.shape)

# -------------------------------
# 4️⃣ Plot temperature slices
# -------------------------------
slice_idx = nz // 2  # middle z-slice
plt.figure(figsize=(8,5))
plt.contourf(X[:,:,slice_idx], Y[:,:,slice_idx], T[:,:,slice_idx], 20, cmap='inferno')
plt.colorbar(label='Temperature [K]')
plt.quiver(X[:,:,slice_idx], Y[:,:,slice_idx], u[:,:,slice_idx], v[:,:,slice_idx], color='white')
plt.xlabel('X [m]')
plt.ylabel('Y [m]')
plt.title('Temperature and Velocity Field (middle Z slice)')
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from src.model import ThermalPINN

# 1️⃣ Load the trained model
model = ThermalPINN(layers=[3, 64, 64, 64, 64, 5], activation='tanh')
model.load_state_dict(torch.load("results/battery_thermal_final.pth"))
model.eval()

# 2️⃣ Generate a 3D grid inside the battery
x = np.linspace(0.0, 0.1, 10)
y = np.linspace(0.0, 0.05, 5)
z = np.linspace(0.0, 0.01, 5)
X, Y, Z = np.meshgrid(x, y, z)
points = np.vstack([X.ravel(), Y.ravel(), Z.ravel()]).T
points_tensor = torch.tensor(points, dtype=torch.float32)

# 3️⃣ Predict temperature and velocity at all points
with torch.no_grad():
    pred = model(points_tensor)
u, v, w, _, T = pred[:,0], pred[:,1], pred[:,2], pred[:,3], pred[:,4]

# 4️⃣ Plot 3D temperature and velocity field
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

# Temperature color
scatter = ax.scatter(X.ravel(), Y.ravel(), Z.ravel(), c=T.numpy(), cmap='inferno', s=50)
fig.colorbar(scatter, ax=ax, label='Temperature [K]')

# Velocity arrows
ax.quiver(X.ravel(), Y.ravel(), Z.ravel(), u.numpy(), v.numpy(), w.numpy(), length=0.01, color='white')

ax.set_xlabel('X [m]')
ax.set_ylabel('Y [m]')
ax.set_zlabel('Z [m]')
ax.set_title('3D Battery Temperature and Velocity Field')

plt.show()
```
